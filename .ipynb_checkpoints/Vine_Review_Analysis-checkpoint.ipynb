```python
import pandas as pd
```

```python
# import vine dataset
vine_df = pd.read_csv('vine_table.csv')
vine_df.head()
```

```python
vine_total_votes = vine_df[vine_df['total_votes'] >= 20]
vine_total_votes.sample(20)
```

```python
vine_helpful = vine_total_votes[vine_total_votes['helpful_votes']/vine_total_votes['total_votes']>=0.5]
vine_helpful.sample(10)
```

```python
vine_paid = vine_helpful[vine_helpful['vine'] == 'Y']
vine_paid.head()
```

```python
vine_unpaid = vine_helpful[vine_helpful['vine'] == 'N']
vine_unpaid.head()
```

```python
total_reviews_paid = len(vine_paid.index)
total_five_star_paid = len(vine_paid[vine_paid['star_rating'] == 5].index)
percent_five_star_paid = total_five_star_paid / total_reviews_paid
```

```python
total_reviews_unpaid = len(vine_unpaid['review_id'].index)
total_five_star_unpaid = len(vine_unpaid[vine_unpaid['star_rating'] == 5].index)
percent_five_star_unpaid = total_five_star_unpaid / total_reviews_unpaid
```

```python
print(f'Total Vine reviews: {total_reviews_paid}\nTotal 5-star Vine reviews: {total_five_star_paid}')
```

```python
print(f'Total non-Vine reviews: {total_reviews_unpaid}\nTotal 5-star non-Vine reviews: {total_five_star_unpaid}')
```

```python
print(f'Percent of Vine reviews with 5 star: {percent_five_star_paid}\nPercent of non-Vine reviews with 5 star: {percent_five_star_unpaid}')
```
